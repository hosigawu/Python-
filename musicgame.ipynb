```python
import pygame
from pygame import mixer
from pygame.locals import *
import time
pygame.init()
clock=pygame.time.Clock()
screen = pygame.display.set_mode((1280,720))
pygame.display.set_caption("千手觀音")
icon = pygame.image.load("千手觀音.jpg")
pygame.display.set_icon(icon)
reddrum=pygame.mixer.Sound("reddrum.mp3")
reddrum.set_volume(0.4)
bluedrum=pygame.mixer.Sound("bluedrum.mp3")
bluedrum.set_volume(0.3)
menudrum=pygame.mixer.Sound("reddrum.mp3")

mixer.music.load("「only my railgun」歌ってみた【＊なみりん】.mp3")
comboword = pygame.font.SysFont('comicsans', 100)
stateword = pygame.font.SysFont('comicsans', 50)
background=pygame.image.load("千手觀音.jpg")
screen.blit(pygame.transform.scale(background, (1280,720)),(0,0))
mixer.music.play()
STARTKEY1=pygame.image.load("START 鍵.jpeg")
screen.blit(pygame.transform.scale(STARTKEY1, (300,50)),(500,580))
#---------------------------------------
pygame.display.update() 
FPS=60
z=False
w=False
per=False
combo=0
comboo=pygame.USEREVENT+1
countt=pygame.USEREVENT+2
number=0
yy=0
gre=False
yy1=0
beat=pygame.image.load("red_circle.png")
beat2=pygame.image.load("blue_circle.png")
yy2=0
miss1=False
pperfect=0
ggreat=0
mmiss=0
running=True

special1=[]
specialcount=0
leftr=False
rightr=False
rightb=False
leftb=False
signal=0
pp=0
forhighcombo=[0]
count=0
sleeplist=[]
musicnode=[]
musicnode2=[]
autoo=False
cho=0
autee=False
def handle_musicnode(musicnode):
    global z
    global w
    global per
    global yy
    global gre
    global yy1
    global yy2
    global miss1
    global combo
    global number
    global specialcount
    global pperfect,ggreat,mmiss
    for musicn2 in musicnode2:
        musicn2.x-=20
        if len(musicnode2)!=0:
            if autee==True:
                if 340<=musicn2.x<=360:
                    yy=20
                    bluedrum.play()
                    pperfect+=1
                    number+=1000
                    per=True
                    musicnode2.remove(musicn2)
                    pygame.event.post(pygame.event.Event(comboo))
                    z=False
            if autee==False:
                if z==True and musicn2.x>490 :
                    z=False
                if (z==True and 460<musicn2.x<=490 ) or musicn2.x<310:
                    yy2=20
                    miss1=True
                    mmiss+=1
                    musicnode2.remove(musicn2)
                    forhighcombo.append(combo)
                    combo=0
                    z=False
                if z==True and 310<=musicn2.x<=460 :
                    if 340<=musicn2.x<=400:
                        yy=20
                        pperfect+=1
                        number+=1000
                        per=True
                        musicnode2.remove(musicn2)
                        pygame.event.post(pygame.event.Event(comboo))
                        z=False
                    if 310<=musicn2.x<340 or 400<musicn2.x<=460:
                        number+=500
                        yy1=20
                        ggreat+=1
                        gre=True
                        musicnode2.remove(musicn2)
                        pygame.event.post(pygame.event.Event(comboo))
                        z=False
    for musicn in musicnode:
        musicn.x-=20
        if len(musicnode)!=0:
            if autee==True:
                if 340<=musicn.x<=360:
                    yy=20
                    reddrum.play()
                    pperfect+=1
                    number+=1000
                    per=True
                    musicnode.remove(musicn)
                    pygame.event.post(pygame.event.Event(comboo))
                    w=False
            if autee==False:
                if w==True and musicn.x>490 :
                    w=False
                if (w==True and 460<musicn.x<=490 ) or musicn.x<310:
                    yy2=20
                    mmiss+=1
                    miss1=True
                    musicnode.remove(musicn)
                    forhighcombo.append(combo)
                    combo=0
                    w=False
                if w==True and 310<=musicn.x<=460 :
                    if 340<=musicn.x<=400:
                        yy=20
                        pperfect+=1
                        number+=1000
                        per=True
                        musicnode.remove(musicn)
                        pygame.event.post(pygame.event.Event(comboo))
                        w=False
                    if 310<=musicn.x<340 or 400<musicn.x<=460:
                        number+=500
                        ggreat+=1
                        yy1=20
                        gre=True
                        musicnode.remove(musicn)
                        pygame.event.post(pygame.event.Event(comboo))
                        w=False
 
def draw_window(musicnode):
    global specialcount
    global per
    global yy
    global gre
    global yy1
    global miss1
    global yy2
    global leftr,rightr,rightb,leftb
    global count
    if count==1:
        if cho==0:
            musicground=pygame.image.load("選歌背景1.png")
            screen.blit(pygame.transform.scale(musicground, (1280,720)),(0,0))
        if cho==1:
            musicground2=pygame.image.load("選歌背景2.png")
            screen.blit(pygame.transform.scale(musicground2, (1280,720)),(0,0))
        if cho==2:
            musicground3=pygame.image.load("選歌背景3.png")
            screen.blit(pygame.transform.scale(musicground3, (1280,720)),(0,0))
        pygame.display.update()
    if count==2:
        if cho==0:
            screen.fill([0,0,0])
            background2=pygame.image.load("six trillion.jpg")
            screen.blit(pygame.transform.scale(background2, (1280,420)),(0,300))
            background3=pygame.image.load("1640507847037.jpeg")
            screen.blit(pygame.transform.scale(background3, (1280,300)),(0,0))
        if cho==1:
                screen.fill([0,0,0])
                background2=pygame.image.load("夜之趨.jpg")
                screen.blit(pygame.transform.scale(background2, (1280,420)),(0,300))
                background3=pygame.image.load("夜之趨鼓背.png")
                screen.blit(pygame.transform.scale(background3, (1280,300)),(0,0))
        if leftb==True:
            leftb=False
            leftbback=pygame.image.load("左半藍.png")
            screen.blit(pygame.transform.scale(leftbback, (118,136)),(147,161))
        if rightb==True:
            rightb=False
            rightbback=pygame.image.load("右半藍.png")
            screen.blit(pygame.transform.scale(rightbback, (118,136)),(147,160))
        if rightr==True:
            rightr=False
            rightrback=pygame.image.load("右半紅.png")
            screen.blit(pygame.transform.scale(rightrback, (118,136)),(147,160))   
        if leftr==True:
            leftr=False
            leftrback=pygame.image.load("左半紅.png")
            screen.blit(pygame.transform.scale(leftrback, (118,136)),(147,160))

        targetzone=pygame.image.load("target_area.png")
        screen.blit(pygame.transform.scale(targetzone, (130,105)),(350,168))
        targetzone2=pygame.image.load("next背.jpeg")
        screen.blit(pygame.transform.scale(targetzone2, (60,105)),(300,168))
        combotext = comboword.render("Combo: " + str(combo), 1, (0,225,0))
        screen.blit(combotext, (600, 10))
        combotext = stateword.render("score :" + str(number), True, (0,0,255))
        screen.blit(combotext, (230, 60))
        if miss1==True or yy2!=0:
            yy1=0
            yy=0
            yy2-=1
            perftext = stateword.render("miss " , True, (0,225,0))
            screen.blit(perftext, (500, 290))
            miss1=False
        
        if (per==True or yy!=0)and (max(yy,yy1)==yy or yy==yy1):
            yy1=0
            yy-=1
            perftext = stateword.render("perfect " , True, (0,225,0))
            screen.blit(perftext, (500, 290))
            peruse=pygame.image.load("yellowcircle.png")
            screen.blit(pygame.transform.scale(peruse,(160,120)),(335,160))
            per=False  
        if (gre==True or yy1!=0 )and max(yy,yy1)==yy1 and yy!=yy1:
            yy1-=1
            perftext = stateword.render("great " , True, (0,225,0))
            screen.blit(perftext, (500, 290))
            greuse=pygame.image.load("greencircle.png")
            screen.blit(pygame.transform.scale(greuse,(130,108)),(347,164))
            gre=False

        for musicn2 in musicnode2:
            screen.blit(pygame.transform.scale(beat2, (75,75)),(musicn2.x,musicn2.y))
        for musicn in musicnode:
            screen.blit(pygame.transform.scale(beat, (75,75)),(musicn.x,musicn.y))
        pygame.display.update()
def file1(file2):
    
    one=open(file2+'.csv',"r")
    for line in one.readlines(): 
        line2 = line.strip()
        line2=line2[:]
        line2=line2.split(" ") 
        l=line2[1]
        if line2[2]=="d":
            if cho==0:  
                special1.append(int(line2[0]))
            if cho==1:
                special1.append(int(line2[0]))
        if cho==0:
            sleeplist.append(float(l)-1.08)#25 0.93 20 1.08
        if cho==1:
            sleeplist.append(float(l)-1.12)#15 1.9 20 1.2 25 0.9 30 
    global endtime0
    endtime0=sleeplist[-1]
def end(ending,perfect,great,miss,combo,score):
    global running
    screen.fill((150,190,230))
    if combo==0:
        combo=pperfect+ggreat
    p=["perfect:",str(pperfect)]
    g=["great:",str(ggreat)]
    m=["miss:",str(mmiss)]
    c=["combo:",str(combo)]
    s=["score:",str(score)]
    font = pygame.font.SysFont("comicsans", 25, True, (0, 0, 0))
    text_clicked = font.render("Clicked", True, (0,0,0))
    textscore1 = font.render(''.join(p), True, (0, 0, 0))
    text_rect1 = text_clicked.get_rect(center=(640, 50))
    textscore2 = font.render(''.join(g), True, (0, 0, 0))
    text_rect2 = text_clicked.get_rect(center=(640, 100))
    textscore5 = font.render(''.join(m), True, (0, 0, 0))
    text_rect5 = text_clicked.get_rect(center=(640, 150))
    textscore3 = font.render(''.join(c), True, (0,0,0))
    text_rect3 = text_clicked.get_rect(center=(640, 200))
    textscore4 = font.render(''.join(s), True, (0,0,0))
    text_rect4 = text_clicked.get_rect(center=(640, 250))
    screen.blit(textscore1, text_rect1)
    screen.blit(textscore2, text_rect2)
    screen.blit(textscore5, text_rect5)
    screen.blit(textscore3, text_rect3)
    screen.blit(textscore4, text_rect4)
    pygame.display.update()
    if escapetime>=endtime0+15:
        pygame.display.quit()
        running=False
        



while running:
    clock.tick(FPS)
    for event in pygame.event.get():
        if event.type==pygame.QUIT:
            running=False
        if event.type == KEYDOWN :
            if event.key == ord('s') and pp==0:
                menudrum.play()
                pp=1
                count+=1
            if count==1 and event.key == ord('d') and cho!=2:
                menudrum.play()
                cho+=1
            if count==1 and event.key == ord('a') and cho!=0:
                menudrum.play()
                cho-=1
            if count==1 and event.key == ord('x') :
                autoo=True
                menudrum.play()
                if cho==0:
                    file1("sixyear")
                    mixer.music.load("六兆年太鼓.ogg")
                    mixer.music.play()
                    background3=pygame.image.load("1640507847037.jpeg")
                    screen.blit(pygame.transform.scale(background3, (1280,300)),(0,0))
                    starttime=time.time()
                    count+=1
                if cho==1:
                    file1("night")
                    mixer.music.load("夜之趨2.ogg")
                    mixer.music.play()
                    background3=pygame.image.load("夜之趨鼓背.png")
                    screen.blit(pygame.transform.scale(background3, (1280,300)),(0,0))
                    starttime=time.time()
                    count+=1
                if cho==2:
                    pass
            if count==1 and event.key == pygame.K_RETURN:
                menudrum.play()
                if cho==0:
                    file1("sixyear")
                    mixer.music.load("六兆年太鼓.ogg")
                    mixer.music.play()
                    background3=pygame.image.load("1640507847037.jpeg")
                    screen.blit(pygame.transform.scale(background3, (1280,300)),(0,0))
                    starttime=time.time()
                    count+=1
                if cho==1:
                    file1("night")
                    mixer.music.load("夜之趨2.ogg")
                    mixer.music.play()
                    background3=pygame.image.load("夜之趨鼓背.png")
                    screen.blit(pygame.transform.scale(background3, (1280,300)),(0,0))
                    starttime=time.time()
                    count+=1
                if cho==2:
                    pass

                
                pygame.display.update()
        if count==2 : 
            if event.type == comboo:
                combo += 1
            if autoo==True:
                autee=True
            if event.type==KEYDOWN:
                if event.key == K_ESCAPE:
                        running=False
            if autoo==False:
                if event.type==KEYDOWN:    
                    if event.key == ord('f') or event.key == ord('j'):
                        if event.key == ord('f'):
                            leftr=True
                        if event.key == ord('j'):
                            rightr=True
                        reddrum.play()
                        w=True
                    if event.key == ord('d') or event.key == ord('k'):
                        if event.key == ord('d'):
                            leftb=True
                        if event.key == ord('k'):
                            rightb=True
                        bluedrum.play()
                        z=True  
                    if event.key == K_ESCAPE:
                        running=False
    if count==2:  
        escapetime=time.time()-starttime
        yaya=[x for x in sleeplist if x <= escapetime]
        for x in yaya:
            sleeplist.remove(x)
            specialcount+=1
            if specialcount in special1:
                musicn2=pygame.Rect(1280,177,10,10)
                musicnode2.append(musicn2)
            else:
                musicn=pygame.Rect(1280,177,10,10)
                musicnode.append(musicn)
    if count==0 or count==1 or count==2 and escapetime<endtime0+5:
        handle_musicnode(musicnode)
        draw_window(musicnode) 
    if count==2 and escapetime>=endtime0+5:
        end(True,perfect=pperfect,great=ggreat,miss=mmiss,combo=max(forhighcombo),score=number)
```
